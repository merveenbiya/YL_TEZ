```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

# Fiili Geçiş - Öğle - Öğrenci

```python
data = pd.read_csv('/content/drive/MyDrive/Merve Enbiyaoğlu Tez/data/data_fiiligecis_ogle_ogrenci.csv')

etiket = data['fiili_gecis_ogrenci_ogle']

# Kategorik sütunları sayısal formata
label_encoder = LabelEncoder()
data['corba'] = label_encoder.fit_transform(data['corba'])
data['ana_yemek'] = label_encoder.fit_transform(data['ana_yemek'])
data['ara_yemek'] = label_encoder.fit_transform(data['ara_yemek'])
data['ek_gida'] = label_encoder.fit_transform(data['ek_gida'])
data['haftanin_günleri'] = label_encoder.fit_transform(data['haftanin_günleri'])
data['dst_donemi'] = label_encoder.fit_transform(data['dst_donemi'])
data['mevsim'] = label_encoder.fit_transform(data['mevsim'])
data['beklenen_hava_olayi'] = label_encoder.fit_transform(data['beklenen_hava_olayi'])
data['anlik_hava_olayi'] = label_encoder.fit_transform(data['anlik_hava_olayi'])

#print(data)


data = data.drop('fiili_gecis_ogrenci_ogle', axis=1)

model = RandomForestRegressor()

model.fit(data, etiket)

feature_importance = model.feature_importances_

feature_importance_df = pd.DataFrame({'Feature': data.columns, 'Importance': feature_importance})

feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

print(feature_importance_df)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df, hue='Feature', palette='viridis', dodge=False)
plt.title('Fiili Geçiş - Öğle - Öğrenci (Feature Importance)')
plt.show()
```
