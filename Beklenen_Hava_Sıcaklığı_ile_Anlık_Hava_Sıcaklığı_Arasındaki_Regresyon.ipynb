```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv('/content/drive/MyDrive/Merve Enbiyaoğlu Tez/data/data_butun.csv')

x = data['beklenen_hava_sicakligi'].values.reshape(-1, 1)
y = data['anlik_hava_sicakligi'].values

plt.figure(figsize=(10, 6))
sns.scatterplot(x='beklenen_hava_sicakligi', y='anlik_hava_sicakligi', data=data)

model = LinearRegression()
model.fit(x, y)

sns.lineplot(x='beklenen_hava_sicakligi', y=model.predict(x), data=data, color='red', label='Regresyon Çizgisi')

plt.title('Beklenen Hava Sıcaklığı ile Anlık Hava Sıcaklığı Arasındaki Regresyon')
plt.xlabel('Beklenen Hava Sıcaklığı')
plt.ylabel('Anlık Hava Sıcaklığı')
plt.legend()
plt.show()

# R-kare
r_squared = r2_score(y, model.predict(x))
print(f'R-squared: {r_squared}')
```
